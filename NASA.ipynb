```python
import pandas as pd
import matplotlib.pyplot as plt
file_path = "/content/neo.csv"
try:
    nasa_df = pd.read_csv(file_path)
    print("Dataset loaded successfully!")
except FileNotFoundError:
    print(f"File not found at {file_path}. Make sure the path is correct.")
    exit()
print("First 5 rows of the dataset:")
print(nasa_df.head())
print("Summary Statistics:")
print(nasa_df.describe())
print("Missing Values:")
print(nasa_df.isnull().sum())
print("Column Names:")
print(nasa_df.columns)
selected_columns = ['id', 'name', 'absolute_magnitude', 'orbiting_body']
selected_data = nasa_df[selected_columns]
print("Selected Data:")
print(selected_data.head())
sorted_data = nasa_df.sort_values(by='absolute_magnitude', ascending=False)
print("Sorted Data (Descending order of absolute magnitude):")
print(sorted_data.head())
plt.hist(nasa_df['absolute_magnitude'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Absolute Magnitude')
plt.xlabel('Absolute Magnitude')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()


```
